## Required Packages

```python
#Import the necessary libraries
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt

sb.set_context("notebook", font_scale=2.5) # Comment this line if you are running directly in Python.
%matplotlib inline
```

## Load the Data

```python
# Loading into Pandas DataFrame
oo_sensory = pd.read_csv("olive_oil_sensory.csv")
```

```python
# Loading into Pandas DataFrame
oo_chemical = pd.read_csv("olive_oil_chemical.csv")
```

## Combining Data

```python
combined = pd.merge(oo_sensory, oo_chemical, how='left', on='region')
combined.head(16)
```

## Delete the first Column

```python
numeric = combined.drop('region', axis=1)
cMatrix = numeric.corr()
sb.heatmap(cMatrix, linewidths=0.5, linecolor="black")
```

## Describe

```python
numeric.describe().T
```

## Boxplot

```python
plt.figure(figsize=(8, 6))
sb.boxplot(data=numeric)
```

## Pairplot

```python
sb.pairplot(numeric)
```

## Histogram

```python

numeric.hist("s_yellow")
```

## Mean, Median, Diff

```python
means = numeric.mean()
print(means)
```

```python
median = numeric.median()
print(median)
```

```python
for i in range(11):
    print(means.iloc[i] - median.iloc[i])
```
